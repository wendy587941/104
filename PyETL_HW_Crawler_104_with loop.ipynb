```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import random
import os

if not os.path.exists('./104jobs'):
    os.mkdir('./104jobs')

url = 'https://www.104.com.tw/jobs/main/'

browser=webdriver.Chrome()
browser.get(url)

browser.find_element_by_xpath('//*[@id="ikeyword"]').send_keys('資料分析師')
browser.find_element_by_xpath('/html/body/article[1]/div[1]/div/div[4]/div/button').click()

for i in range(20): 
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(random.randint(1, 5))
    
k = 1
while k != 0:
    try:
        browser.find_elements_by_class_name("js-more-page",)[-1].click() 
        # 如果真的找不到，也可以直接找中文!
        # driver.find_element_by_xpath("//*[contains(text(),'手動載入')]").click()
        print('Click 手動載入，' + '載入第' + str(15 + k) + '頁')
        k = k+1
        time.sleep(random.randint(3, 5)) # 時間設定太短的話，來不及載入新資料就會跳錯誤
    except:
        k = 0
        print('No more Job')

res=browser.page_source
soup=BeautifulSoup(res, 'html.parser')

for jobUrlTags in soup.select('a[class="js-job-link"]'):
    try:
        jobUrl='https:'+jobUrlTags['href']
        browser.get(jobUrl)
        jobRes=browser.page_source
        jobSoup=BeautifulSoup(jobRes, 'html.parser')
        compName=jobSoup.select('a[class="btn-link t3 mr-6"]')[0]['title']
        jobTitle=jobSoup.select('h1[class="pr-6 text-break"]')[0]['title']
        salary=jobSoup.select('p[class="t3 mb-0 mr-2 monthly-salary text-primary font-weight-bold float-left"]')[0].text
        workType=jobSoup.select('p[class="t3 mb-0"]')[0].text
        location=jobSoup.select('p[class="t3 mb-0"]')[1].text
        weekOff=jobSoup.select('p[class="t3 mb-0"]')[5].text
        workExp=jobSoup.select('p[class="t3 mb-0"]')[8].text
        major=jobSoup.select('p[class="t3 mb-0 major"]')[0].text
        with open('./104jobs/{}.txt'.format(compName), 'w', encoding='utf-8') as f:
            f.write('公司名稱: '+compName+'\n')
            f.write('工作職稱: '+jobTitle+'\n')
            f.write('工作待遇: '+salary+'\n')
            f.write('工作性質: '+workType+'\n')
            f.write('上班地點: '+location+'\n')
            f.write('休假制度: '+weekOff+'\n')
            f.write('工作經驗: '+workExp+'\n')
            f.write('科系要求: '+major+'\n')
        try:
            language=jobSoup.select('p[class="t3 m-0"]')[0].text
            with open('./104jobs/{}.txt'.format(compName), 'a', encoding='utf-8') as f:
                f.write('語文條件: '+language+'\n')
        except IndexError:
            language=jobSoup.select('p[class="t3 mb-0 w-100"]')[0].text
            with open('./104jobs/{}.txt'.format(compName), 'a', encoding='utf-8') as f:
                f.write('語文條件: '+language+'\n')
        try:
            jobSkills=jobSoup.select('p[class="mb-0 r3"]')[1].text
            with open('./104jobs/{}.txt'.format(compName), 'a', encoding='utf-8') as f:
                f.write('工作技能: '+jobSkills+'\n')
        except IndexError:
            with open('./104jobs/{}.txt'.format(compName), 'a', encoding='utf-8') as f:
                f.write('工作技能: '+'不拘'+'\n')
        print('完成輸出一個職缺')        
    except IndexError:
        print('失敗輸出一個職缺!')
    except FileNotFoundError:
        print('失敗輸出一個職缺!')
    time.sleep(random.randint(1, 5))
```

```python
import pandas as pd
import glob

data = list()
for filePath in glob.glob('./104jobs/*.txt'):
    with open(filePath, 'r', encoding='utf-8') as f:
        tmpList = f.read().split('\n')
        data.append(tmpList)
columns = [r.split(': ') [0] for r in data[0]]

df = pd.DataFrame(data=data, columns=columns)

for c in df:
    df[c] = df[c].apply(lambda x: x.split(': ')[-1])
df
```

```python
df.to_csv('./104jobs.csv', index=False, encoding='utf-8-sig')
```
